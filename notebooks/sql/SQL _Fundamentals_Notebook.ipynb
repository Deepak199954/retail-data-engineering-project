Ticket 3 — Highest Value Orders

Finance wants to inspect the top 20 highest-value orders.

Return:

order_id
customer_id
order_amount

```sql
select order_id, customer_id, order_amount from orders order by order_amount desc limit 20
```

Ticket 4 — Newest Customers

Marketing wants to see customers ordered from newest signup date to oldest.

```sql
select customer_id, customer_name, signup_date from customers order by signup_Date desc
```

Operations wants to review pending orders ordered from highest order amount to lowest.

```sql
select order_id, customer_id, order_amount, order_status from orders where order_status = "PENDING" order by order_amount desc
```

#DE-006 — GROUP BY

Ticket 1 — Orders by Status

Display the total number of orders for each order status.

```sql
select count(*), order_status from orders group by order_status
```

Ticket 2 — Revenue by Category

Calculate the total revenue generated by each product category.

Display the highest revenue category first.

```sql
select sum(order_amount), category from orders group by category order by sum(order_amount) desc
```

Ticket 3 — Average Order Value by Category

Calculate the average order amount for each category.

```sql
select avg(order_amount), category from orders group by category
```

Ticket 4 — Customers by City

Display the total number of customers in each city.

Display cities having the highest customer count first.

```sql
select count(*), city from customers group by city order by count(*) desc
```

Ticket 5 — Highest Order by Category

Find the largest order amount recorded for each category.

```sql
select max(order_amount), category from orders group by category
```

Ticket 6 — Missing Revenue by Status

Display the number of orders with missing order_amount grouped by order status.

```sql
select count(*),order_status from orders where order_amount is null group by order_status
```

#DE-007 — HAVING

Ticket 1 — High Revenue Categories

Display categories whose total revenue is greater than 10,000.

```sql
select sum(order_amount) as total_revenue, category from orders group by category having total_revenue > 10000
```

Ticket 2 — Busy Cities

Display cities having more than 20 customers.

```sql
select count(*) as customer_count, city from customers group by city having customer_count > 20
```

Ticket 3 — Frequently Ordered Products

Display products having more than 30 orders.

```sql
select count(*) as total_orders,product_name from orders group by product_name having total_orders > 30
```

Ticket 4 — Premium Categories

Display categories whose average order amount is greater than 800.

```sql
select avg(order_amount) as average_order_amount, category from orders group by category having average_order_amount > 800
```

Ticket 5 — Data Quality Dashboard

Display only those order statuses that have at least one missing order_amount.

```sql
select count(*) as total_count, order_status from orders where order_amount is null group by order_status having total_count >= 1
```
