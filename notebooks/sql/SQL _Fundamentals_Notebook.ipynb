#DE-008 — Business Rules using CASE

Ticket 1 — Order Value Classification
Business Requirement

Create a new column named order_value_category.

Business Rules:

order_amount >= 1000 → High Value
Otherwise → Regular

Return:

order_id
order_amount
order_value_category

```sql
select 
    order_id, 
    order_amount,
    case
        when order_amount >= 1000 then "High Value"
        else "Regular"
    end as order_value_category 
from orders
```

Ticket 2 — Order Processing Status
Business Requirement

Create a new column named processing_priority.

Business Rules:

COMPLETED → Closed
PENDING → Action Required
CANCELLED → No Action
NULL → Unknown

Return:

order_id
order_status
processing_priority

```sql
select 
    order_id, 
    order_status,
    case
        when order_status = "COMPLETED" then "Closed"
        when order_status = "PENDING" then "Action Required"
        when order_status = "CANCELLED" then "No Action"
        else "Unknown"
    end as processing_priority
 from orders
```

Ticket 3 — Quantity Classification
Business Requirement

The warehouse wants to classify orders based on quantity.

Business Rules:

Quantity >= 5 → Bulk Order
Otherwise → Standard Order

Return:

order_id
quantity
order_type

```sql
select 
    order_id, 
    quantity,
    case 
        when 
            quantity >= 5 then "Bulk Order"
        else 
            "Standard Order"
    end as order_type
from orders
```

Ticket 4 — Revenue Band
Business Requirement

Finance wants to categorize orders into revenue bands.

Business Rules:

Less than 500 → Low
500–1000 → Medium
Greater than 1000 → High

Return:

order_id
order_amount
revenue_band

```sql
select 
    order_id, 
    order_amount,
    case
        when order_amount < 500 then "LOW"
        when order_amount >= 500 and order_amount < 1000 then "Medium Greater"
        else "High"
    end as revenue_band
from orders
```

Ticket 5 — Data Quality Indicator
Business Requirement

The Data Quality team wants to quickly identify records with missing revenue.

Create a column called data_quality_status.

Business Rules:

order_amount IS NULL → Missing Amount
Otherwise → Valid Amount

Return:

order_id
order_amount
data_quality_status

```sql
select
    order_id,
    order_amount,
    case
        when order_amount is null then "Missing Amount"
        else "Valid Amount"
    end as data_quality_status
from orders
```

Business Requirement

The Customer Success team wants a report showing every order along with the customer's details.

Return:

order_id
customer_name
city
product_name
order_amount
order_status

Important:

Use the orders and customers Bronze tables.
Use table aliases (o and c).
Don't worry about filtering yet.

```sql
select 
    o.order_id, 
    c.customer_name, 
    o.product_name, 
    o.order_amount, 
    o.order_status 
from 
    orders o 
        inner join customers c  on o.customer_id = c.customer_id
```

#DE-011 — JOIN Practice (Production Scenarios)

Ticket 1 — Customer Order Report
Business Requirement

The Customer Success team wants a report of all customers along with any orders they have placed.

Customers who have never placed an order should still appear.

Return:

customer_id
customer_name
city
order_id
product_name

```sql
select 
    c.customer_id, 
    c.customer_name, 
    c.city, 
    o.order_id, 
    o.product_name, 
    o.order_amount 
from 
    customers c 
        left join 
    orders o 
        on c.customer_id = o.customer_id
```

Ticket 4 — Data Quality Check
Business Requirement

The Data Quality team wants to identify orders that reference customers that do not exist.

Return:

order_id
customer_id
order_amount

This is a real data engineering scenario.

You'll probably need one extra condition after the join.


```sql
select
    o.order_id,
    c.customer_id,
    o.order_amount
from orders o
        left join
    customers c
        on o.customer_id = c.customer_id
    where o.customer_id is null
```
